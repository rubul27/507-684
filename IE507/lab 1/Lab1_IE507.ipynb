## `1. while` loop

`while` loop runs as long as the given condition provided is true. In the above code-snippet, the `while` loop runs as long as `n>0`. After each iteration of the loop, the value of n decreases by 1. When `n=0`, the condition in the `while` loop is not met and we exit from the loop.

```python
n = 6
while n>0:
  print('countdown: ',n)
  n=n-1
print('countdown complete. We have exited the while loop')
```

what happens if we forget to decrease n? Try the below code, and fix error.

```python
n = 5
while n>0:
  print('countdown: ',n)

print('countdown complete. We have exited the while loop')
```

### Infinite while loop
A common mistake when using `while` loop is either
* to forget specifying the exit condition
* to forget updating (or updating it incorrectly) the variable which determines the exit condition
In the above example, since we are not updating the variable, the `while` loop keeps on running unless we force stop it.

We may also use the `break` command to stop a loop. Consider the following example using `break` which replicates out initial finite countdown loop.

```python
n = 5
while True:
  if n <= 0:
    break
  print('countdown:',n)
  n = n-1

print('countdown complete. We have exited the while loop.')
```

### An example code snippet using `while` loop

```python
# This program prints all the divisors a positive integer inputted by the user
n = input("Enter any positive integer: ")
num = int(n)
i = 2
while i<num:
  if num%i == 0:
    print(i, 'is a divisor of ', n)
  i += 1  # this is same as i=i+1
```

### Sum of sequence of natural numbers using `for` loop


Let us write a program to find the sum of first `n` natural numbers where `n` is provided by user.

```python
n = input('Enter n: ')
num = int(n)
sum = 0 # initialization
for i in range(num+1):
  sum += i # same as sum = sum + i
print('The sum of first ',num,' natural numbers is ',sum)
```

### Factorial example

```python
n = input('Enter n: ')
num = int(n)
prod = 1 # initialization
for i in range(1,num+1):
  prod = prod * i
print(str(num)+'! = '+str(prod))
```

## 5. Fibonacci series

Recall, a Fibonacci series is mathematically represented as $F_0 = 0$, $F_1 = 1$ and for $n \geq 2$, $F_n = F_{n-1} + F_{n-2}$. So the second Fibonacci number is $F_2 = F_0 + F_1 = 0 + 1 = 1$. The third Fibonacci number is $F_3 = F_1 + F_2 = 1 + 1 = 2$ and so on.

```python
print ('Program to output first 20 Fibonacci numbers')
a = 0
b = 1
maxcount = 20
for i in range(1, maxcount+1):
    print (i, 'th number is ', a)
    temp = a + b
    a = b
    b = temp
```

We can write the code more compactly using `list`. We will use `append` functionaity of a `list`. `append` is used to add new elements to the end of the list.

```python
a = [1,2,3]
a.append(4)
a
```

Let's now write the code for generating Fibonacci numbers using `list`

```python
n=5
fibsequence = [0,1]
for i in range(2,n):
  fibsequence.append(fibsequence[i-1]+fibsequence[i-2])
print(fibsequence)
```

Let us go through above code snippet line by line.
*  First, we define the variable n, and set it to the integer 20.
  * n is the length of the sequence we're going to form.

* We then create a variable called fibsequence, and initialize it to the list with the integers 0 and 1 in it, the first two elements of the Fibonacci sequence. We have to create these elements "by hand", since the iterative part of the sequence requires two previous elements.

* We then have a for-loop over the list of integers from 2 (the next element of the list) to n (the length of the sequence).
    * Try and see what happens when we set n < 2.   

* In the body of the loop, we append to the list an integer equal to the sum of the two previous elements of the list.

* After exiting the loop (ending the indentation) we then print out the whole list. That's it!

## 6. Functions

Below module is an example for functions used in python. To define a function we need `def` keyword followed by name to specify function and arguments for function.  

1) `compare(var1, var2)` used to compare two numbers and return bigger number using conditional statements

2) `Print_table(number)` function used to print table of a number by calling another function `table()`.

```python
a = 5
b = 7

def compare(var1, var2):
  if var1 < var2:
    return (str(var2) + " is bigger")
  elif var1> var2:
    return (str(var1) + " is bigger")
  else:
    return("Both are equal numbers")
```

```python
compare(a,b)
```

```python
def table(number):
  temp = []
  for i in range(1, 11):
    temp.append(i*number)
  return(temp)


def Print_table(number):
  for i in range (1, number+1):
    print(table(i))
```

```python
Print_table(5)
```

We might want to use the Fibonacci snippet with different sequence lengths. It will be easier and more useful to make a function out of the code. We do this with the `def` statement in Python.

```python
def fibonacci(sequence_length):
    "Get Fibonacci sequence of length *sequence_length*"
    fibsequence = [0,1]
    if sequence_length < 1:
        print("Fibonacci series defined for length >= 1")
        return
    if 0 < sequence_length < 3:
        return fibsequence[:sequence_length]
    for i in range(2,sequence_length):
        fibsequence.append(fibsequence[i-1]+fibsequence[i-2])
    return fibsequence
```

```python
fibonacci(5)
```

First, note that the function itself is defined as a code block (a colon followed by an indented block). Next, note that the first line of the function is a single string. This is called a docstring, and is a special kind of comment that is used to document what the function does.
