```python
def time_convert(t):
    # Convert a time string in "hh:mm" format to minutes since midnight
    hours, minutes = map(int, t.split(':'))
    total_minutes = hours * 60 + minutes
    return total_minutes

def reconvert(minutes):
    # Convert minutes since midnight back to "hh:mm" format
    minutes = minutes % (24 * 60)
    hours, minutes = divmod(minutes, 60)
    time_string = f"{hours:02d}:{minutes:02d}"
    return time_string

def add_time(t1, t2):
    # Add two time strings in "hh:mm" format and return the result
    total_minutes = time_convert(t1) + time_convert(t2)
    return reconvert(total_minutes)
```

```python
def maximum_time(t1, t2):
    s1 = time_convert(t1)
    s2 = time_convert(t2)
    if s2 >= s1:
        return t2
    else:
        return t1

feasible_trains = []
for s1_train_no, s1_train_arr in enumerate(ARR.loc['S1', 'M1']):
    for s2_train_no, s2_train_arr in enumerate(ARR.loc['S2', 'M1']):
        if maximum_time(s1_train_arr, s2_train_arr) == s1_train_arr:
            move_over_time = time_convert(reconvert(time_convert(s1_train_arr) + 60))
        else:
            move_over_time = time_convert(reconvert(time_convert(s2_train_arr) + 60))

        for m1_train_no, m1_train_dep in enumerate(DEP.loc['M1', 'C1']):
            if time_convert(m1_train_dep) > move_over_time:
                feasible_trains.append([s1_train_no, s2_train_no, m1_train_no])

    feasibles = []
    prev0, prev1, prev2 = feasible_trains[0]
    sublist = [prev2]

    for a, b, c in feasible_trains[1:]:
        if (a == prev0) and (b == prev1):
            sublist.append(c)
        else:
            feasibles.append([prev0, prev1, copy.deepcopy(sublist)])
            sublist.clear()
            sublist.append(c)
        prev0 = a
        prev1 = b

    layover_trains = len(feasibles)

print("Number of layover trains:", layover_trains)
print("Feasible train combinations:", feasible_trains)
```

```python
def time_difference(t1, t2):
    # Calculate the absolute time difference between two time values
    return abs(time_convert(t1) - time_convert(t2))

def find_feasible_trains(ARR, DEP, layover_time, wait_time):
    feasible_trains = []
    for s1_train_no, s1_train_arr in enumerate(ARR.loc['S1', 'M1']):
        for s2_train_no, s2_train_arr in enumerate(ARR.loc['S2', 'M1']):
            max_time_train = max(s1_train_arr, s2_train_arr)
            if time_difference(s1_train_arr, s2_train_arr) <= wait_time:
                move_over_time = time_convert(reconvert(time_convert(max_time_train) + layover_time))
                for m1_train_no, m1_train_dep in enumerate(DEP.loc['M1', 'C1']):
                    if time_convert(m1_train_dep) > move_over_time:
                        feasible_trains.append([s1_train_no, s2_train_no, m1_train_no])
    return feasible_trains

def group_feasible_trains(feasible_trains):
    feasibles = []
    prev0, prev1, prev2 = feasible_trains[0]
    sublist = [prev2]
    for a, b, c in feasible_trains[1:]:
        if a == prev0 and b == prev1:
            sublist.append(c)
        else:
            feasibles.append([prev0, prev1, copy.deepcopy(sublist)])
            sublist.clear()
            sublist.append(c)
        prev0 = a
        prev1 = b
    return feasibles

def create_heatmap(feasibles, layover_time, wait_time):
    h = np.zeros((len(ARR.loc['S1', 'M1']), len(ARR.loc['S2', 'M1'])))
    g = [i for i in range(len(ARR.loc['S1', 'M1']))]
    k = [i for i in range(len(ARR.loc['S2', 'M1']))]
    df = pd.DataFrame(h, index=g, columns=k)

    for a, b, ls in feasibles:
        df.loc[a][b] = len(ls)

    sns.heatmap(df)
    plt.xlabel("Train Number from S2 to M1")
    plt.ylabel('Train Number from S1 to M1')
    plt.title(f"Feasible Trains from M1 to C1 for {layover_time / 60}hr Layover and {wait_time / 60}min Waiting Time")
    plt.show()

layover_times = [60, 120, 300]
waiting_times = [90, 150]

for layover_time in layover_times:
    for wait_time in waiting_times:
        feasible_trains = find_feasible_trains(ARR, DEP, layover_time, wait_time)
        feasibles = group_feasible_trains(feasible_trains)
        create_heatmap(feasibles, layover_time, wait_time)
```
