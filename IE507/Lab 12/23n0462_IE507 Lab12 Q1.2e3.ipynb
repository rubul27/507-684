```python
from sklearn.cluster import KMeans
```

```python
X = df1.to_numpy()
```

2. For the data in S1.txt, vary the number of clusters K by choosing from the set {6, 7, 8, 9, 10, 11, 12, 13}
and use the data in the KMeans function of scikit learn package

```python
def km(n):
  km = KMeans(n_clusters = n, random_state=0)
  km_result = km.fit(X)
  km_result.labels_
  print('For number of cluster = ',n)

  sns.scatterplot(x=X[:,0],y=X[:,1],hue=km_result.labels_, palette='pastel')
  plt.scatter(x=km_result.cluster_centers_[:,0],y=km_result.cluster_centers_[:,1],color='black',marker="D", s=34)
  plt.show()
```

4. Consider the test.txt given above and find the predictions for the points in test.txt for clustering
obtained for each value of K. Plot the points in the scatter plot and indicate the predicted cluster
labels.


```python
df1test = pd.read_csv('test.txt',sep = '    ',index_col=None,header = None)
df1test
```

```python
new_points = df1test.to_numpy()
```

```python
markers=["p","<","*","h",">","^","s","o"]
for n in range(3,20):
  km = KMeans(n_clusters = n, random_state=0)
  km_result = km.fit(X)
  km_result.predict(new_points)
  sns.scatterplot(x=X[:,0],y=X[:,1],hue=km_result.labels_, palette='bright')
  plt.show()

  for j in range(len(new_points)):
    plt.scatter(x=new_points[j,0],y=new_points[j,1],color= 'black',marker=markers[j], s=44 )
```

6.Using your procedure to find the best choice of number of clusters derived in Problem 2.1, report the
best choice for the number of clusters

```python
wcss = []

for i in range(3, 20,2):
    clustering = KMeans(n_clusters=i, init='k-means++', random_state = 0)
    clustering.fit(X)

    wcss.append(clustering.inertia_)

ks = [ 3,5, 7, 9, 11,13,15,17,19]

sns.lineplot(x = ks, y = wcss )
```

In the graph , it is seen that the elbow is formed at the cluster points 5 or 7. So according to the elbow method , the ideal number of clusters can be 5 or 7
