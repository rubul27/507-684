```python
import numpy as np
import pandas as pd
import matplotlib as plt
import numpy.random as rnd
```

```python
temp = np.array([[1000,1150,1300,1450,1600],[0.15,0.30,0.10,0.25,0.20]])
```

```python
df= pd.DataFrame(temp,index = ['Bearing life(in hrs)','Probability'])
```

```python
df
```

```python
delaytime = rnd.triangular(5,10,15)
newbearinginstallcharge = 320   #per bearing
downtimecost = 100 #per min
onsitecost = 300  #perhr
installationtime = 20 #in minutes


time_taken = []
for i in range(100):
  timetakentoreplace = delaytime * installationtime * 2 # for changing two bearings
  time_taken.append(timetakentoreplace)

print("Expected replacing time :",np.mean(time_taken),'minutes')
print(time_taken)
```

```python
mean_bearinglife = 0
for num in list(temp[0]*temp[1]):
  mean_bearinglife += num
```

```python
mean_bearinglife
```

```python
bearing_hrs = 10000
t = 0
n = 0            #no of times repaired
while t <= bearing_hrs:
  t += mean_bearinglife +np.mean(time_taken)
  n +=1


```

```python
total_time_in_reparing = n * np.mean(time_taken)
```

```python
total_time_in_reparing
```

```python
total_cost = total_time_in_reparing * (newbearinginstallcharge*2) * (downtimecost)*(onsitecost/60)
```

```python
total_cost
```
