```python
import numpy as np
import matplotlib.pyplot as plt

def hinge_loss(z):
    return np.maximum(0, 1 - z)

def logistic_loss(z):
    return np.log(1 + np.exp(-z))

def squared_hinge_loss(z):
    return np.maximum(0, 1 - z)**2

# Generate values for z
z_values = np.linspace(-10, 10, 1000)

# Calculate loss values
hinge_values = hinge_loss(z_values)
logistic_values = logistic_loss(z_values)
squared_hinge_values = squared_hinge_loss(z_values)

# Plot the loss functions
plt.figure(figsize=(12, 6))
plt.plot(z_values, hinge_values, label='Hinge Loss', color='blue')
plt.plot(z_values, logistic_values, label='Logistic Loss', color='red')
plt.plot(z_values, squared_hinge_values, label='Squared Hinge Loss', color='green')
plt.xlabel('z')
plt.ylabel('Loss')
plt.title('Comparison of Hinge, Logistic, and Squared Hinge Loss Functions')
plt.legend()
plt.grid(True)
plt.show()
```
