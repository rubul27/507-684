**1. What is the minimizer and minimum function value of $f(x) = f(x_1, x_2) = (a + 1 − x_1)^2 + b · (x_2 − x_1^2)^2 $?,
Where $\frac{b−10}{10}$ is the last digit of your Roll Number and a is 1 if b is an even number, else a is 2. Use these
values a, b for other problems that follows.**

Roll no = 23n0462. So, b=30 and a = 1
Therefore the equation becomes
 $f(x) = f(x_1, x_2) = (2 − x_1)^2 +30 (x_2 − x_1^2)^2 $

```python
import numpy as np
from numpy.linalg import inv
from matplotlib import ticker, cm

def f(x):
    return (2 - x[0])**2 + 30 * (x[1] - x[0]**2)**2

def grad_f(x):
    return np.array([ -2 * (2 - x[0]) - 120 * x[0] * (x[1] - x[0]**2), 60 * (x[1] - x[0]**2)])

def hessian_f(x):
    return np.array([ [2 + 240 * x[0]**2 - 120 * (x[1] - x[0]**2), -120 * x[0]],[-120 * x[0], 60]])

def newtons_method(f, grad_f, hessian_f, x0, tolerance):
    x = x0
    while np.linalg.norm(grad_f(x)) > tolerance:
        hessian_inv = inv(hessian_f(x))
        x = x - hessian_inv @ grad_f(x)
    return x

x0 = np.array([-2, 2])  # Initial point
tolerance = 1e-6  # Tolerance level

minimizer = newtons_method(f, grad_f, hessian_f, x0, tolerance)

print(f"Minimizer: {minimizer}")
print(f"Minimum function value: {f(minimizer)}")
```

**3. With the starting point $x_0 = (−1.5, 1.5)$, we aim to analyze the behavior of the algorithm 3 for different tolerance
values. We set $τ = 10−p$ where p = 1, 2, . . . , 20. For each τ , record the final minimizer, objective function value
at termination, and the number of iterations required for convergence in a tabular form. Generate a plot,
illustrating the relationship between the number of iterations and τ values. Comment on the observations.
Comment about the minimizers and objective function values obtained for different choices of the tolerance
values. Plot the level sets of the function in (1) and also plot the trajectory of the optimization on the same
plot and report your observations. ”In optimization, a trajectory refers to the path or sequence of points that
a numerical optimization algorithm traverses while iteratively updating the solution in search of an optimal
point”.**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (2 - x[0])**2 + 30 * (x[1] - x[0]**2)**2

def grad_f(x):
    return np.array([
        -2 * (2 - x[0]) - 120 * x[0] * (x[1] - x[0]**2),
        60 * (x[1] - x[0]**2)
    ])

def hessian_f(x):
    return np.array([
        [2 + 240 * x[0]**2 - 120 * (x[1] - x[0]**2), -120 * x[0]],
        [-120 * x[0], 60]
    ])

def newtons_method(f, grad_f, hessian_f, x0, tolerance):
    x = x0
    iterations = 0
    while np.linalg.norm(grad_f(x)) > tolerance:
        hessian_inv = np.linalg.inv(hessian_f(x))
        x = x - hessian_inv @ grad_f(x)
        iterations += 1
    return x, f(x), iterations

x0 = np.array([-1.5, 1.5])  # Initial point
tolerances = [10**-p for p in range(1, 21)]  # Tolerance values

minimizers = []
function_values = []
num_iterations = []

for tolerance in tolerances:
    minimizer, function_value, iterations = newtons_method(f, grad_f, hessian_f, x0, tolerance)
    minimizers.append(minimizer)
    function_values.append(function_value)
    num_iterations.append(iterations)

# Print the results in a tabular form
print("Tolerance\tFinal Minimizer\tObjective Function Value\tIterations")
for i in range(len(tolerances)):
    print(f"{tolerances[i]}\t{minimizers[i]}\t{function_values[i]}\t{num_iterations[i]}")

# Generate a plot of the relationship between the number of iterations and τ values
plt.figure(figsize=(10, 6))
plt.semilogx(tolerances, num_iterations, marker='o')
plt.title('Number of Iterations vs. Tolerance')
plt.xlabel('Tolerance (τ)')
plt.ylabel('Number of Iterations')
plt.grid(True)
plt.show()
```
