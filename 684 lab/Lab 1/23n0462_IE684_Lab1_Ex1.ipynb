1. $f(x_1, x_2) = 100x_2^2 + 100x_1^4 - 200x_2 x_1^2 + x_1^2 - 2x_1 + 1$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import ticker, cm
```

```python
# Define the function
def f1(x1, x2):
    return 100 * x2**2 + 100 * x1**4 - 200 * x2 * x1**2 + x1**2 - 2 * x1 + 1

# Create a grid of x1 and x2 values
x1 = np.linspace(-4, 4, 100)
x2 = np.linspace(-4, 4, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f1(X1, X2)

# Plot the level sets
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1, X2, Z, locator = ticker.LogLocator())
plt.colorbar(contour, shrink=0.8, extend='both')
plt.show()
```

2. $f(x_1,x_2) = x_1^4 + x_2^4 - 20x_1^3 - 20x_2^3 + 100x_1^2 +100x_2^2$

```python
# Define the function
def f2(x1, x2):
    return x1**4 +x2**4 - 20 * x1**3 - 20 * x2**3 + 100*x1**2 + 100*x2**2
# Create a grid of x1 and x2 values
x1 = np.linspace(-10, 20, 100)
x2 = np.linspace(-10, 20, 100)
X1, X2 = np.meshgrid(x1, x2)
Z2 = f2(X1, X2)

# Plot the level sets
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1, X2, Z2,locator = ticker.LogLocator())
# plt.clabel(contour, inline=True, fontsize=8)
# plt.xlabel('x1')
# plt.ylabel('x2')
# plt.title('Level Sets of f2(x1, x2)')
plt.colorbar(contour, shrink=0.8, extend='both')
plt.show()
```

3. $f(x_1,x_2) = -cos(x_1)cos(x_2)exp(-((x_1-\Pi)^2 + (x_2 - \Pi)^2))$

```python
# define the function
def f3(x1, x2):
    return -np.cos(x1) * np.cos(x2) * np.exp(-((x1 - np.pi)**2 + (x2 - np.pi)**2))

# Create a grid of x1 and x2 values
x1 = np.linspace(0,5, 300)
x2 = np.linspace(0,5, 300)
X1, X2 = np.meshgrid(x1, x2)
Z3 = f3(X1, X2)

# Plot the level sets
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1,X2,Z3)
plt.colorbar(contour, shrink=0.8, extend='both')
plt.show()
```

4.  $ f(x_1,x_2) =  2x_1^2 - 1.05x_1^4 + \frac{x_1^6}{6}  + x_1x_2 + x_2^2 $

```python
# Define the function
def f4(x1, x2):
    return 2*x1**2  - 1.05 * x1**4 + x1**6/6 +x1*x2 +x2**2
# Create a grid of x1 and x2 values
x1 = np.linspace(-4,4, 100)
x2 = np.linspace(-4, 4, 100)
X1, X2 = np.meshgrid(x1, x2)
Z4 = f4(X1, X2)

# Plot the level sets
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1, X2, Z4,locator = ticker.LogLocator())
plt.colorbar(contour, shrink=0.8, extend='both')
plt.show()
```
