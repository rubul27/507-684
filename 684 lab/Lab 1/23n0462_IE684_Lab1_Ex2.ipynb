```python
import math

# Function f(x) = √(x^2 + 5)
def f(x):
    return math.sqrt(x**2 + 5)

# Search space S = {x : −0.2− 2 ≤ x ≤ 0.2 + 2}
search_space = (-2.2, 2.2)

# Values of ε to test
epsilons = [ 10**(-2), 10**(-3), 10**(-4), 10**(-5)]

# Calculate the grid resolution and grid points X for each ε
for epsilon in epsilons:
    # Calculate grid resolution
    grid_resolution = 2 * epsilon / 1  # L = 1

    # Generate grid points X
    X = [x for x in np.arange(int(search_space[0] / grid_resolution), int(search_space[1] / grid_resolution) + 1)]

    # Evaluate f(x) for all x in X
    f_values = [f(x * grid_resolution) for x in X]

    # Find the minimum x that minimizes f(x)
    min_x = X[f_values.index(min(f_values))]

    # Output the results
    print(f"For ε = {epsilon}:")
    print(f"Grid resolution = {grid_resolution}")
    print(f"Minimum x that minimizes f(x): {min_x }")
    print(f"Objective function values at x: {min([round(val, 5) for val in f_values])}")
    print()
```

```python
import numpy as np

def f2(x1, x2):
    return np.sqrt(x1**2 + x2**2)

# Define the search space
x1_range = np.arange(-2.2, 2.3, 0.1)
x2_range = np.arange(-2.4, 2.5, 0.1)
X1, X2 = np.meshgrid(x1_range, x2_range)
```

```python
epsilons = [10**(-1), 2*10**(-1), 5*10**(-1), 10**(-2), 2*10**(-2), 5*10**(-2), 10**(-3)]
L = 1
minima = []

for epsilon in epsilons:
    # Calculate the grid resolution
    grid_resolution = 2 * epsilon / L

    # Generate the grid points
    X = np.column_stack((X1.ravel(), X2.ravel()))

    # Evaluate the function at each grid point
    evaluations = f2(X[:, 0], X[:, 1])

    # Find the minimum value and its corresponding point
    min_index = np.argmin(evaluations)
    min_point = X[min_index]
    min_value = evaluations[min_index]

    minima.append((epsilon, min_point, min_value))

# Print the results
for epsilon, min_point, min_value in minima:
    print(f"For epsilon = {epsilon}, the minimum point is {min_point} with a value of {min_value}")
```

```python
import matplotlib.pyplot as plt

# Plot the function
plt.contourf(X1, X2, f2(X1, X2), levels=50, cmap='viridis')
plt.colorbar(label='f(x1, x2)')
plt.xlabel('x1')
plt.ylabel('x2')

# Plot the minima
for _, min_point, _ in minima:
    plt.plot(min_point[0], min_point[1], 'ro')

plt.title('Level Set of f(x1, x2) with Minima')
plt.show()

```
